# Visualize the correlation matrix with seaborn 

```
sns.heatmap(data2.corr(), annot= True)
plt.show()
```

# Discount and Profit

```
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Profit', data = data2, color = 'r', label= 'Discount')
plt.legend()
```

## From the lineplot of Profit vs Discount we can conclude that there is a negative association between these two variables i.e. if the value of discount increases, the value of profit decreases

# Discount and Sales

```
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Sales', data = data2, color = 'r', label= 'Discount')
plt.legend()
```

## From the lineplot of Sales vs Discount we can conclude that there is a negative association between these two variables i.e. if the value of discount increases, the value of profit decreases. The value of sales is highest when discount is given at rate 50% then the amount of sales decreases drastically.

# Discount and Quantity

```
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Quantity', data = data2, color = 'r', label= 'Discount')
plt.legend()
```

## From the lineplot of Quantity vs Discount we can say that for all the discounts the Quantity purchased is overall same.

# Histogram of the data

```
data2.hist(bins=50 ,figsize=(20,15))
plt.show()
```

## From the histogram of the whole data, we can conclude that, the distribution of Sales, Quantity and Discount is positively skewed and only the Profit is normally distributed.

# Now we will Group or sum the sales ,profit,discount,quantity according to every state of region and also according to sub-categories sales

```
grouped=pd.DataFrame(data2.groupby(['Ship Mode','Segment','Category','Sub-Category','State','Region'])['Quantity','Discount','Sales','Profit'].sum().reset_index())
grouped
```

# sum,mean,min,max,count median,standard deviation,variance of each states of Profit

```
data2.groupby("State").Profit.agg(["sum","mean","min","max","count","median","std","var"])
```
