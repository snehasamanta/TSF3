# Discount and Sales

```
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Sales', data = data, color = 'r', label= 'Discount')
plt.legend()
```

## From the lineplot of Sales vs Discount we can conclude that there is a negative association between these two variables i.e. if the value of discount increases, the value of profit decreases. The value of sales is highest when discount is given at rate 50% then the amount of sales decreases drastically.

```
sales_profit=data.groupby(['Sub-Category'])['Profit','Sales'].agg(['sum'])
sales_profit.plot.bar(rot=0,figsize=(20,10))

plt.title('Sales Profit per Sub-Category',fontsize=35)
plt.show(block=True)
```

```
sales_profit=data.groupby(['Category'])['Profit','Sales'].agg(['sum'])
sales_profit.plot.bar(rot=0,figsize=(20,10))

plt.title('Sales Profit per Category',fontsize=35)
plt.show(block=True)
```

```
discount_profit=data.groupby(['Sub-Category'])['Discount','Profit'].agg(['sum'])
discount_profit.plot.bar(rot=0,figsize=(20,10))

plt.title('Discount Profit per Sub-Category',fontsize=35)
plt.show(block=True)
```

```
discount_sales=data.groupby(['Sub-Category'])['Discount','Sales'].agg(['sum'])
discount_sales.plot.bar(rot=0,figsize=(20,10))

plt.title('Discount Sales per Sub-Category',fontsize=35)
plt.show(block=True)
```

```
discount_sales=data.groupby(['Category'])['Discount','Sales'].agg(['sum'])
discount_sales.plot.bar(rot=0,figsize=(20,10))

plt.title('Discount Sales per Category',fontsize=35)
plt.show(block=True)
```

```
sales_profit=data.groupby(['Segment'])['Profit','Sales'].agg(['sum'])
sales_profit.plot.bar(rot=0,figsize=(20,10))

plt.title('Sales Profit per Segment',fontsize=35)
plt.show(block=True)
```

```
discount_profit=data.groupby(['Segment'])['Discount','Profit'].agg(['sum'])
discount_profit.plot.bar(rot=0,figsize=(20,10))

plt.title('Discount Profit per Segment',fontsize=35)
plt.show(block=True)
```
